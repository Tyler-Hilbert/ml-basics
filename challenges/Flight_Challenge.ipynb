Read csv

```python
import pandas as pd
df = pd.read_csv('flights.csv')
df.sample(10)
```

Check for null values

```python
df.isnull().sum()
```

# Take a closer look at the null values

```python
df_NaNs = df[df.isnull().any(axis=1)]
df_NaNs
```

DepDel15, "a binary indicator that departure was delayed by more than 15 minutes (and therefore considered late)", is the only row with null/NaN values.




```python
df_NaNs.groupby('DepDelay').first()
```

All DepDel15 have a DepDelay of 0, so we can set DepDel15 to 0.
We should check if there are any values where the delay>15minutes&DepDel==0.0 or delay<=15minutes&DepDel==0.0

```python
df.groupby('DepDel15').first()
```

```python
computed_depdel15 = pd.Series(df.DepDelay > 15)
df = pd.concat([df, computed_depdel15.rename("ComputedDepDel15")], axis=1)
df
```

```python
df.groupby(['DepDel15', 'ComputedDepDel15']).first()
```

```python
df.groupby(['DepDel15', 'ComputedDepDel15']).DepDelay.count()
```

```python
df[df.DepDel15==1.0][df.ComputedDepDel15==False].groupby('DepDelay').first()
```

# Fill null values and continue

```python
df.DepDel15 = df.DepDel15.fillna(0) # I checked and all the NaN DepDel15 have no delay
df
```

# Look at the data

```python
df.describe()
```

```python
departure_cutoff = df.DepDelay.quantile(.997)
arrival_cutoff = df.ArrDelay.quantile(.997)

df = df[df.DepDelay < departure_cutoff]
df = df[df.ArrDelay < arrival_cutoff]
```

```python
from matplotlib import pyplot as plt

fix, ax = plt.subplots(1, 2, figsize=(16, 4))

dep_delay = df.DepDelay
ax[0].hist(dep_delay)
ax[0].set_title("Departure Delay")
ax[0].set_xlabel("Minutes")
ax[0].set_ylabel("Count")

arr_delay = df.ArrDelay
ax[1].hist(arr_delay)
ax[1].set_title("Arrival Delay")
ax[1].set_xlabel("Minutes")
ax[1].set_ylabel("Count")

plt.show()
```
